```python
#크롤링
 
#유튜브 영상을 크롤링해보자
#시범으로 ikon의 고무줄다리기를 다운받아보자 
import os
import subprocess
import pytube
 
yt = pytube.YouTube("https://www.youtube.com/watch?v=PfgbZr4tFU0") #다운받을 동영상 URL 지정
 
vids= yt.streams.all()
 
#영상 형식 리스트 확인
for i in range(len(vids)):
    print(i,'. ',vids[i])
 
vnum = int(input("다운 받을 화질은? "))
 
 
parent_dir = "C:/Users/kebee/Desktop"
vids[vnum].download(parent_dir) #다운로드 수행
 
new_filename = input("변환 할 mp3 파일명은?")
 
default_filename = vids[vnum].default_filename 
subprocess.call(['ffmpeg', '-i',                 #cmd 명령어 수행
    os.path.join(parent_dir, default_filename),
    os.path.join(parent_dir, new_filename)
])
 
print('동영상 다운로드 및 mp3 변환 완료!')
 
 
 
#full 영화를 크롤링 해보자
#영화들을 딥러닝으로 19세영화와 그 외의 영화로 자동 분리하는 시스템을 만드는 것이 목적이다 
#먼저 유튜브 영화영상 url부터 따보자 
 
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pandas import DataFrame
import numpy as np
import time
driver = webdriver.Chrome('C:/Users/kebee/Desktop/python/chromedriver.exe') 
driver.get('https://www.youtube.com/results?search_query=movie+full') # 셀레니움으로 크롬을 조정해 'movie full' 검색한 유튜브 링크로 들어간다
html = driver.page_source 
soup = BeautifulSoup(html, 'html.parser') #html을 따온다 
title_movie_all= []
url_full_all= []
i = 0
while i < 8: # 스크롤을 내리지 않으면 더 많은 영화를 볼 수 없으므로 스크롤을 내리면서 url을 딴다  
    i = i+1
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") 
    time.sleep(1)
    for link in soup.select('h3 > a'):
        url_tail = link.get('href') #url 꼬리부분 크롤링 
        url_full = 'https://www.youtube.com' + url_tail #url 합치기 
        url_full_all.append(url_full)
        title_movie = link.get('title') #영상 제목 따기 
        title_movie_all.append(title_movie)
 
 
 
#이제는 실제 영화를 크롤링한다 
import os
import subprocess
import pytube
 
for idx, url in enumerate(url_full_all):
 
    yt = pytube.YouTube(url) 
    vids= yt.streams.all()
 
    vnum = 0 #화질
 
 
    parent_dir = "C:/Users/kebee/Desktop/movie"
    vids[vnum].download(parent_dir) #다운로드 수행
 
    new_filename = title_movie_all[idx] # 영화 제목 지정 
 
    default_filename = vids[vnum].default_filename 
    subprocess.call(['ffmpeg', '-i',                 #cmd 명령어 수행
    os.path.join(parent_dir, default_filename), # 저장
    os.path.join(parent_dir, new_filename) # 저장 
])
 
 
 
driver.close()
```
