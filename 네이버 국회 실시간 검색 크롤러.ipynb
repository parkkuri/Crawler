```python
from selenium import webdriver
import os
import pandas as pd
import numpy as np
import pickle
import time
os.chdir('C:/Users/q/workspace/council')

 
browser = webdriver.Firefox()
#국회
browser.get('https://search.naver.com/search.naver?where=realtime&sm=tab_jum&query=%EA%B5%AD%ED%9A%8C')

count_all =[]
count = 1
all = []
while True:
    try:
        menus = browser.find_elements_by_css_selector('#realTimeSearchBody > div.rt_wrap.rt_typ2 > div > ul > li.first_child > dl > dd > a')
        menus2 = browser.find_elements_by_css_selector('#realTimeSearchBody > div.rt_wrap.rt_typ2 > div > ul > li > dl')
        for inx in range(len(menus2)):
            content = menus2[inx].text.split('\n')[1]
            try:
                url = menus[0].get_attribute('href')
            except:
                url = menus[0].get_attribute('data-src')
            time = ' '.join(menus2[inx].text.split('\n')[2].split(' ')[0:2])
            single= list([content,url,time])
            print(str(count)+'번째 성공')
            count +=1
            all.append(single)
            count_all =[]
            
        try:
            import time

            browser.execute_script("window.scrollTo(0, 10000000);")
            time.sleep(3)
            browser.find_elements_by_css_selector('#realTimeSearchBody > div.rt_wrap.rt_typ2 > div.bt_more > a')[0].click()
            time.sleep(2)
            print('루프 돌기')
        except:
            break
#         if count>500:
#             article = pd.DataFrame(all,columns=['content','url','time'])
#             with open('국회'+str(count)+'.txt', 'wb') as f:
#                 pickle.dump(article,f)
        
        
    except:
        print(str(count)+'번째 실패')
        count_all.append(count)
        count +=1
        if len(count_all)>5:
            break
        continue

article = pd.DataFrame(all,columns=['content','url','time'])
article.to_csv('article_국회3.csv',index=False,encoding='utf16',sep='\t')
```
